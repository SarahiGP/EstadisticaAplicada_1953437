```python
## librerías

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## datos
#1
personas = ['Juan', 'Jorge', 'Ana', 'Alba']
edades= [23, 18, 18, 31]
#2
dic1={'Municipio': ['Escobedo', 'Apodaca', 'Monterrey', 'Guadalupe'], 'Nacimientos': [345, 487, 372, 298], 'Muertes': [25, 56, 39, 67], 'Divorcios': [28, 46, 95, 19]}
data_mun= pd.DataFrame(dic1)
```

```python
data_mun
```

```python
## Gráfica de barras de las edades matplotlib
plt.bar(personas, edades)
plt.title('Edades de desconocidos')
plt.xlabel('Personas')
plt.ylabel('Edad en años')
plt.show()
```

```python
## grafica de barras para los nacimientos matplotlib
plt.bar(data_mun['Municipio'], data_mun['Nacimientos'])
plt.title('Nacimientos en NL')
plt.xlabel('municipio')
plt.ylabel('número de nacimientos')
plt.show()
```

```python
## definir el pizarrón/lienzo
fig, (panel1, panel2) = plt.subplots(2, 1)
#grafica1 nacimientos
#plt.tipo(ejex, ejey)
panel1.bar(data_mun['Municipio'], data_mun['Nacimientos'])
#grafica 2 edades
panel2.bar(personas, edades)
##
plt.show()
```
