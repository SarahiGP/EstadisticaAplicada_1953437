```python
# Librerias básicas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Librerias especializadas
from statsmodels.tsa.seasonal import seasonal_decompose

##Configurar tamaño de las graficas (largo, alto)
plt.rcParams['figure.figsize'] = (10 ,6)
```

```python
#datos
pasageros = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/airline_passengers.csv',index_col = 'Month', parse_dates= True)
pasageros.head()
```

```python
pasageros = pasajeros.rename(columns= {'Thousands of Passengers' : 'Miles de Pasageros'})
```

```python
pasageros['Miles de Pasageros'].plot()
plt.title('Pasageros de avion E.U 1949-1960')
plt.xlabel('Meses')
plt.ylabel('Miles');
```

1. La serie tiene una tendencia creciente, no tiene ciclo, tienen componente estacional cada verano de cada año se incrementa el numero de pasajeros.

Vamos a obtener los componentes de la serie:
1. Tendencia-ciclo (trend)
2. Estacional (seasonal)
3. irregular (error)

```python
## error-trend-seasonal
ets1 = seasonal_decompose(pasageros['Miles de Pasageros'], model = 'mult')
```
